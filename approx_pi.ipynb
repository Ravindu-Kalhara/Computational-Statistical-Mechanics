# Approximation of $\pi$ using Monte Carlo Simulations

```python
import numpy as np
import matplotlib.pyplot as plt
```

<img src="./imgs/approx_pi.png" alt="Approximation of Pi" width="400"/>   

The Monte Carlo method can be used to approximate $\pi$ using the following relation.   
$\pi = 4 \times \frac{\text{Number of points in the Circle}}{\text{Number of points in the Square}}$

```python
def approx_pi(a, num_iter, num_samples):
    num_pts_circle = 0
    approx_pi_arr = np.zeros(num_samples)
    divisor = num_iter // num_samples

    for iter in range(num_iter):
        x = a * np.random.random()
        y = a * np.random.random()

        r = x*x + y*y
        if r < a*a:
            num_pts_circle += 1

        if iter % divisor:
            approx_pi_arr[iter//divisor] = 4 * (num_pts_circle/iter)

    return approx_pi_arr
```

```python
a = 1
num_iter = 2000000
num_samples = 50 # number of samplings done throughout the iteration.
x_arr = (num_iter//num_samples) * np.arange(num_samples)

approx_pi_arr = approx_pi(a, num_iter, num_samples)
```

```python
plt.figure(figsize=(10, 5))
plt.plot(x_arr, approx_pi_arr, marker="o", markersize=3)
plt.axhline(y=np.pi, linestyle="--", color="r")
plt.title(r"Approximation of $\pi$")
plt.xlabel("Number of iterations")
plt.ylabel("Approximation")
plt.show()
```
